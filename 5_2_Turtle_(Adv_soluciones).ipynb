*Challenge 8:* Dibuja dos estrellas de 50 puntas:

```
initializeTurtle()
speed(11)
# Dibujar la estrella de 50 picos
size = 200  # Tamaño de los lados de la estrella
angle = 360 / 50  # Ángulo para girar y dibujar cada pico

for _ in range(0,2):
  for _ in range(50):
      forward(size)
      right(144)

      forward(size)
      right(144 - angle)
  right(180)
```

*Challenge 9:* Dibuja la siguiente figura:

```
initializeTurtle()
speed(11)
# Dibujar la espiral con 3 brazos
radius = 10  # Radio inicial del brazo
angle = 120  # Ángulo para girar en cada paso
sides = 60  # Número de pasos para dibujar la espiral

for _ in range(sides):
    forward(radius)
    right(angle)
    radius += 5  # Aumentar el radio en cada paso para crear la espiral
```
