```python
from random import randint

#crea una lista de opciones
t = ["verde", "rojo", "azul"]

#assigna un color al azar para la vaiable "elección"
eleccion = t[randint(0,2)]
print(eleccion)
```

```python
def elevar_al_cuadrado(n):
   valor = n*n
   return valor

num = float(input("Enter a number: "))
print(elevar_al_cuadrado(num))
```

```python
def tabla_del(numero):
  lista=[]
  for i in range(1,11):
    lista.append(numero*i)
  return lista
```

```python
tabla_del(8)
```

```python
def multiplicar(x,y=2):
   result = x*y
   return result
print(multiplicar(2,3))
```
